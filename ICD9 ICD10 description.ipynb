```python
import requests
import bs4
import re
import csv
import xlsxwriter
import nltk
from nltk.tokenize import word_tokenize
```

```python
icd9=['icd9']
icd10=['icd10']
icd9_des=['icd9_des']
icd10_des=['icd10_des']
```

```python
dx9=["263.9","278.01","262","263.0","263.1","263.8","278.00","278.0","278.03"]

for j in dx9:
    print(j)
    
    res1 = requests.get('https://www.findacode.com/code.php?set=ICD9&c='+j)
    soup1 = bs4.BeautifulSoup(res1.text, 'lxml')
    description1_obj = soup1.select('title')
    if description1_obj:description1_text=description1_obj[0].getText()
    
    res2 = requests.get('http://icd9.chrisendres.com/index.php?srchtype=diseases&srchtext='+j+'&Submit=Search&action=search')
    soup2 = bs4.BeautifulSoup(res2.text, 'lxml')
    description2_obj = soup2.select('div.dlvl')
    if description2_obj:description2_text=description2_obj[0].getText()
    
    
    if "Medical Codes" in description1_text:
        regexp = re.compile(j+"\s([\w\d\s$&+,:;=?@#|'<>.^*()%!-]+)")
        description_regx=regexp.search(description2_text).group(1)
    else:
        regexp = re.compile("[\w\d\.]+\s([\w\d\s$&+,:;=?@#|'<>.^*()%!-]+)\s\-\sICD")
        description_regx=regexp.search(description1_text).group(1)
        
    print(description_regx)
    icd9.append(j)
    icd9_des.append(description_regx)
```

```python
dx10=["E66.9","E46","E66.09","E66.01","E66.1","E43","E44.1","E66.8","E44.0","E66.2","O99.215","E64.0","P05.2","E40","E41","E42","E43","E44.0","E44.1","E45","E46","E64.0","R64","E66.01","E66.2","Z68.41","Z68.42","Z68.43","Z68.44","Z68.45"]

for i in dx10:
    print(i)
    
    res = requests.get('https://www.findacode.com/code.php?set=ICD10CM&c='+i)
    soup = bs4.BeautifulSoup(res.text, 'lxml')
    description_obj = soup.select('title')
    description_text=description_obj[0].getText()
    regexp = re.compile("[\w\d\.]+\s([\w\d\s$&+,:;=?@#|'<>.^*()%!-]+)\s\-\sICD")
    description_regx=regexp.search(description_text).group(1)

    print(description_regx)
    icd10.append(i)
    icd10_des.append(description_regx)
```

```python
rows = zip(icd9,icd9_des)

with open('icd9.csv', 'w',newline='') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```

```python
rows = zip(icd10,icd10_des)

with open('icd10.csv', 'w',newline='') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
